## Dataset .txt

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dinisoraya/uts-pdt/No_1/main/CellPhone_train.txt')
df.head()
```

```python
df.info()
```

```python
column_sum = df.sum()
column_mean = df.mean()
column_median = df.median()

stats_df = pd.DataFrame({
    'Sum': column_sum,
    'Mean': column_mean,
    'Median': column_median
})

stats_df
```

## Dataset .csv

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/dinisoraya/uts-pdt/No_1/main/Coffee_Qlty.csv')
df2.head()
```

```python
df2.info()
```

```python
numerical_columns = df2.select_dtypes(include=['float64', 'int64']).columns
```

```python
column_sum = df2[numerical_columns].sum()
column_mean = df2[numerical_columns].mean()
column_median = df2[numerical_columns].median()

stats_df = pd.DataFrame({
    'Sum': column_sum,
    'Mean': column_mean,
    'Median': column_median
})

stats_df
```
